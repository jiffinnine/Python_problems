```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# I am reading all the clean files for analysis
df_hindi = pd.read_csv('cleaned_movies_data/cleaned_bollywood_movies.csv')
df_telugu = pd.read_csv('cleaned_movies_data/cleaned_tollywood_movies.csv')
df_indian = pd.read_csv('cleaned_movies_data/cleaned_indian_movies.csv')
```

```python
# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
avg_genre = round(df_all.groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre
```

```python
avg_genre.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'All Indian movies Genre avg box office')
```

## Inferences
- From this we can understand that in Indian cinema, the top p=erforming genres are biographical sports drama, fantasy, action and science fiction, mostly due to mass appeal and entertainment.
- On the other hand the least performing genres are Sports Drama, sports action, Thriller, family drama , which all tends to be more critical thinking than entertainment.

```python
avg_genre_bolly = round(df_hindi[df_hindi['Language'] == 'Hindi'].groupby('Genre')['BoxOffice (Crores)'].mean(), 2).reset_index()
```

```python
avg_genre_bolly.sort_values('BoxOffice (Crores)', ascending = False).plot.barh(x= 'Genre', title = 'Bollywood Genre avg box office')
```

## Inferences
- From this we can see that there are less genres in Bollywood indicating less variety.
- We can also see that biographical sports drama is a hit even in bollywood.
- There is a lot more interest in comedy, action and romance compared to the all India industry.
- Lowest performing is fmaily drama, thriller since more interest is given to historical and biographical elements in movie-making
