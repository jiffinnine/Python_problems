```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# I am reading all the clean files for analysis
df_hindi = pd.read_csv('cleaned_movies_data/cleaned_bollywood_movies.csv')
df_telugu = pd.read_csv('cleaned_movies_data/cleaned_tollywood_movies.csv')
df_indian = pd.read_csv('cleaned_movies_data/cleaned_indian_movies.csv')
```

```python
# Made a dataframe wich merges all 3 dfs to find top 3 across all languages
df_all = pd.concat([df_hindi, df_telugu, df_indian], ignore_index = True)
```

```python
df_all['BoxOffice (Crores)'].mean()
```

```python
df_all['BoxOffice (Crores)'].plot(kind='hist', bins=10, edgecolor = 'Black', title='Box Office Distribution')
```

```python
plt.boxplot(df_all['BoxOffice (Crores)'])
plt.title('BoxOffice (Crores)')
plt.ylabel('Revenue')
plt.show()
```

## Inferences
- We can derive from this histogram that most percentage of movies belong in the box office category of < 300 Crores.
- From the boxplot we can see that most movies are in the 0 - 300 range and also the orange line indicates the median to be around 296.
- The outliers are the floating circles indicating that few movies have had over the top collections.
